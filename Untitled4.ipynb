```python
A = np.array([
    [2, 3],
    [1, 4]
    ])

B = np.array([
    [0, -1],
    [3, -7]
])

C = A*B

print(C)
print("\n")

D = A@B
print(D)
```

```python
A = np.array([
    [1, 2, 3],
    [2, 2, 4],
    [1, 5, -1]
    ])

B = np.array([
    [3, 4, 5],
    [1, 7, 6],
    [0, 2, 1]
])

C = A@B

print(C)

D = A@A

print("\n")
print(D)

E = A**2

print("\n")
print(E)

F = B@A

print("\n")
print(F)

```
